```python
# Configurações de conexão
cluster = Cluster(['localhost'], port=9042)
session = cluster.connect()
```

```python
session.execute("""
        CREATE KEYSPACE IF NOT EXISTS imdb
        WITH REPLICATION = {
            'class' : 'SimpleStrategy',
            'replication_factor' : 1
        }
    """)
```

```python
session.set_keyspace('imdb')
```

```python
# Lista de comandos CQL para criar tabelas
commands = [
    """
    CREATE TABLE IF NOT EXISTS imdb.cf_genero (
        id_genero UUID PRIMARY KEY,
        nome_genero text
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS imdb.cf_elenco (
        id_elenco UUID PRIMARY KEY,
        nome_elenco text
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS imdb.cf_escritor (
        id_escritor UUID PRIMARY KEY,
        nome_escritor text
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS imdb.cf_diretor (
        id_diretor UUID PRIMARY KEY,
        nome_diretor text
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS imdb.cf_filme (
    id_filme UUID PRIMARY KEY,
    nome_titulo_filme text,
    ano_lancamento_filme text,
    indicado_posicao_ranking_filme UUID,
    classificacao_indicativa text,
    duracao text,
    orcamento text,
    bilheteria text,
    slogan text
);
    """
]
```

```python
# Executando cada comando CQL individualmente
for command in commands:
    session.execute(command)
```

```python
# Configurações do CSV
csv_file_genero = 'C:\\Users\\jess1\\Desktop\\Apache-Cassandra\\dados\\cf_genero.csv'
csv_file_elenco = 'C:\\Users\\jess1\\Desktop\\Apache-Cassandra\\dados\\cf_elenco.csv'
csv_file_diretor = 'C:\\Users\\jess1\\Desktop\\Apache-Cassandra\\dados\\cf_diretor.csv'
csv_file_filme = 'C:\\Users\\jess1\\Desktop\\Apache-Cassandra\\dados\\cf_filme.csv'
csv_file_escritor = 'C:\\Users\\jess1\\Desktop\\Apache-Cassandra\\dados\\cf_escritor.csv'
keyspace_name = 'imdb'
table_name_genero = 'cf_genero'
table_name_elenco = 'cf_elenco'
table_name_diretor = 'cf_diretor'
table_name_filme = 'cf_filme'
table_name_escritor = 'cf_escritor'
```

```python
# Ler o arquivo CSV e inserir dados na tabela
with open(csv_file_genero, 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file, delimiter=';')
    next(csv_reader)  # Pula o cabeçalho se houver
    query = session.prepare(f"INSERT INTO {table_name_genero} (id_genero, nome_genero) VALUES (?, ?);")
    for row in csv_reader:
        print(row)  # Adicione esta linha para imprimir os dados do CSV
        uuid_id = uuid.uuid4()  # Gera um UUID aleatório
        session.execute(query, (uuid_id, row[1]))
```

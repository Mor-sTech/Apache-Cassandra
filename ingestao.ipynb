```python
from cassandra.cluster import Cluster
from cassandra.auth import PlainTextAuthProvider
import csv
import uuid
```

```python
# Configurações de conexão
cluster = Cluster(['localhost'], port=9042)
session = cluster.connect()
```

```python
session.execute("""
        CREATE KEYSPACE IF NOT EXISTS imdb
        WITH REPLICATION = {
            'class' : 'SimpleStrategy',
            'replication_factor' : 1
        }
    """)
```

```python
session.set_keyspace('imdb')
```

```python
#escrevendo nosso comando de criação para a tabela filme
command = """
CREATE TABLE IF NOT EXISTS imdb.cf_genero (
    id_genero UUID PRIMARY KEY,
    nome_genero text
);
"""
```

```python
session.execute(command)
```

```python
# Configurações do CSV
csv_file = 'C:\\Users\\jess1\\Desktop\\Apache-Cassandra\\dados\\cf_genero.csv'
keyspace_name = 'imdb'
table_name = 'cf_genero'
```

```python
# Ler o arquivo CSV e inserir dados na tabela
with open(csv_file, 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file, delimiter=';')
    next(csv_reader)  # Pula o cabeçalho se houver
    query = session.prepare(f"INSERT INTO {table_name} (id_genero, nome_genero) VALUES (?, ?);")
    for row in csv_reader:
        print(row)  # Adicione esta linha para imprimir os dados do CSV
        uuid_id = uuid.uuid4()  # Gera um UUID aleatório
        session.execute(query, (uuid_id, row[1]))
```
